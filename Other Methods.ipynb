## Other methods

Using `scipy` we have an easy time fitting to whatever function we want, but we have to define the function first. This can be sped up by building a quick package that includes these fitable functions like gaussians or voigtians but I havent done that. 

Instead we can use numpy to fit data using their polyfit. This function can give polynomial fits for a given degree. So in order to get our fits for power curves we need to do a log plot of some sort.

```python
try:
    import numpy as np; import scipy as sc; import matplotlib.pyplot as plt; import uncertainties
except ImportError:
    import sys
    !{sys.executable} -m pip install numpy scipy matplotlib pathlib uncertainties
    print("Restart your kernel and try again")
```

```python
# Its a good idea to include all your imports needed to run the cell since you dont have to re-run cells all the time
import numpy as np; import matplotlib.pyplot as plt

x = np.linspace(start = 0, stop = 5, num = 100) #1,000 Equally spaced points from 0-5 

#initialize our rng with a fixed seed (2) such that every time after this we will get the same data
rng = np.random.default_rng(seed=2)

noise = .25*rng.normal(loc=0, scale=1, size=100) #gaussian distribution centered at x = 0, with variance = 1 used to simulate noise
data = np.abs(5*np.exp(-(x-1)**2/2) + noise) #Simulated data abs value for future use

# rng.normal?
#If you forget what a function does, and what parameters are needed, you can put a questionmark after the function to return the documentation on it. 
#You can also use tab to bring up a list of possible functions, i.e. rng.(tab) brings up a list of all possible probability distribution functions. 
```

```python
plt.loglog(x,data)
plt.grid()
#plotting on log log show this isn't a power law, so next lets try a logy 
```

```python
from uncertainties import ufloat
plt.semilogy(x,data, 'g.')
# here we see this data could be fit to some concave function. But, after 3, the uncertianty blows out the real data.
#Lets try with a polynomial of degree 2

coef, cov = np.polyfit(x, np.log(data), deg=2 , cov = True)
uncert = np.sqrt(np.diag(cov))
```

We are trying to fit this function with something like 
$$
    y = ae^{bx+cx^2}
$$

since we are fitting on a ylog scale, the polynomial that gets fit will be 

$$
    log(y) = log(a)+bx+cx^2
$$

such that $log(a)$ = **coef[2]**, $b$ = **coef[1]**, and $c$ = **coef[0]**. 

```python
fit = np.exp(np.polyval(coef, x))
plt.plot(x,fit)
plt.plot(x,data,'g.')

a = ufloat(coef[0],uncert[0])
b = ufloat(coef[1],uncert[1])
c = ufloat(coef[2],uncert[2])

print(cov)
print("Amplitude : {:.8}".format(a))
print("Standard Deviation : {:.8}".format(b))
print("Mean : {:.8}".format(c))
```

This method doesnt work very well for non linear plots as we have seen for gaussians. For examples of places where it does work well here is a good [link](https://plainenglish.io/blog/a-guide-to-data-fitting-in-python-3715c815fffd).
